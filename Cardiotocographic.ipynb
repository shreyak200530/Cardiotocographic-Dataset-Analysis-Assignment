```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Cardiotocographic.csv")
```

```python
df.head(100)
```

```python
df.info()
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
sns.boxplot(df['LB'])
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['LB'], bins=20, edgecolor='black')
plt.title("Distribution of Baseline Fetal Heart Rate (LB)")
plt.xlabel("LB (Baseline Fetal Heart Rate)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.violinplot(y=df['LB'], color='skyblue')
plt.title("Violin Plot of LB")
plt.show()
```

```python
Q1 = df['LB'].quantile(0.25)
Q3 = df['LB'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['LB'] >= Q1 - 1.5*IQR) & (df['LB'] <= Q3 + 1.5*IQR)]
```

```python
df.describe()
```

```python
df.median()

```

```python
df.std()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
corr = df.corr()
corr['LB'].sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='AC', y='LB', hue='FM', data=df, palette='viridis')
plt.title("Comparison of LB vs AC (colored by FM)")
plt.xlabel("AC (Accelerations)")
plt.ylabel("LB (Baseline Fetal Heart Rate)")
plt.show()
```
